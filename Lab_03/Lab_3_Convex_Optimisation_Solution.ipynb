# Find the value of $x$ at which $f(x)$ is minimum :

**Example 1** : $f(x)$ = $x^{2}$ + $x$ + $2$

**Analytical** : $$\frac{d}{dx} f(x)=2x+1=0$$
$$\frac{d^{2}}{dx^{2}} f(x)=2~(Minima)$$
$$x=-\frac{1}{2}~(analytical~xution)$$

**Gradient Descent Update equation** : $$x_{init}=4$$
$$x_{updt}=x_{old}-\lambda (\frac{d}{dx} f(x)|x=x_{old})$$
$$x_{updt}=x_{old}-\lambda (2x_{old}+1)$$

**Gradient Descent Method** :

Follow the below steps and write your code in the block below

1. Generate $x$, 1000 data points from -10 to 10
2. Generate and Plot the function $f(x)$ = $x^{2}$ + $x$ + $2$
3. Initialize the starting point ($x_{init}$) and learning rate ($\lambda$)
4. Use Gradient descent algorithm to compute value of $x$ at which the function $f(x)$ is minimum
5. Also vary the learning rate and initialisation point and plot your observations

```
## Write your code here

# anlytical solution
x = Symbol('x')
y = x**2 + x +2
dy = y.diff(x)
print(dy)
z=solve(dy)
dy_dy=dy.diff(x)
print(dy_dy)
w=sympify(dy_dy,z)
if w>0:
  print('\n','The minimum value found at x(by analytical method) = ',z)
else:
  print('\n ERROR')


# to return function value at given point 
def val(a):
  return a**2+a+2

# to return derivative value at ppoint
def der(z):
  return 2*z+1

def gradient_descent(val, der, nlp,lr):
	# track all x_val
	x_val, f_val = list(), list()
	# generate an initial point
	x_old=x_init
	# run the gradient descent
	for i in range(nlp):
		# calculate gradient
		gradient = der(x_old)
		# take a step
		x_old = x_old - lr * gradient
		# evaluate candidate point
		x_old_eval = val(x_old)
		# store x_old
		x_val.append(x_old)
		f_val.append(x_old_eval)
	return [x_val, f_val]

x=np.arange(-10,10,0.02)
y=x**2+x+2
lr=0.8
x_init=10
x_min=0
x_val, f_val = gradient_descent(val,der,30,lr)  
print('Minimum value of function : ', np.amin(f_val))

for i in range(len(x_val)):
  if val(x_val[i])==np.amin(f_val):
    x_min=x_val[i]
    exit() 

print('\nValue of x corresponding minimum value : ',x_min,'\n')
plt.plot(x,y)
plt.plot(x_val,f_val,'.-',color='g')
plt.plot(x_min,np.amin(f_val),'x',color='r')
plt.xlabel('x')
plt.ylabel('f(x)')

plt.show()
```

#Find the value of $x$ and $y$ at which $f(x,y)$ is minimum :



**Example 1** : $f(x,y)$ = $x^{2}+y^{2}+2x+2y$

**Gradient Descent Method** :

Follow the below steps and write your code in the block below

1. Generate $x$ and $y$, 1000 data points from -10 to 10
2. Generate and Plot the function $f(x,y)$ = $x^{2}+y^{2}+2x+2y$
3. Initialize the starting point ($x_{init}$,$y_{init}$) and learning rate ($\lambda$)
4. Use Gradient descent algorithm to compute value of $x$ and $y$ at which the function $f(x,y)$ is minimum
5. Also vary the learning rate and initialisation point and plot your observations

```
## Write your code here
import numpy as np
import matplotlib.pyplot as plt
from sympy import *
from sympy import Eq, Symbol, solve, sin, cos
import math as mt
from mpl_toolkits.mplot3d import Axes3D

# anlytical solution
x = Symbol('x')
y = Symbol('y')
z = x**2 + y**2 + 2*x + 2*y
# first derivatives
p = z.diff(x)
q = z.diff(y)
print('\n',p ,'\n',q)
s1=solve(p)
s2=solve(q)
print(s1,s2)
# second derivatives
r=p.diff(x)
t=q.diff(y)
s=p.diff(y)
print('\n',r,'\n',s,'\n',t)
def d(s1,s2):
  fxx=sympify(r,s1,s2)
  fyy=sympify(t,s1,s2)
  fxy=sympify(s,s1,s2)
  return fxx*fyy-fxy**2
D=d(s1,s2) 
print(D)
if D>0:
  if sympify(r,s1,s2)>0:
    print('\n Minimum value found at (x,y) : (',s1,s2,')\n')
  else:
    print('\n ERROR')



# to return function value at given point 
def val(a,b):
  return a**2+b**2+2*a+2*b

# to return derivative value at ppoint
def der_x(z1):
  return 2*z1+2
def der_y(z2):
  return 2*z2+2

def gradient_descent(val, der_x, der_y, nlp, lr, xi,yi):
  varval=np.empty((0,2))
  f_val=list()
  xo=xi
  yo=yi
  for i in range(nlp):
    gradx=der_x(xo)
    grady=der_y(yo)
    xo=xo-lr*gradx
    yo=yo-lr*grady
    xo_val=val(xo,yo)
    varval = np.append(varval, np.array([[xo,yo]]), axis=0)
    f_val.append(xo_val)
  return [varval, f_val]

X, Y = np.linspace(-10,10,100), np.linspace(-10,10,100)
x, y = np.meshgrid(X,Y)
lr=0.6
xi=5
yi=5
x_min=0
x_val, f_val = gradient_descent(val, der_x, der_y, 30, lr, xi, yi)
 
print('Minimum value of function : ', np.amin(f_val))

for i in range(len(x_val[:,0])):
  for j in range(0,2):
    a=x_val[i][0]
    b=x_val[i][1]
    if val(a,b)==np.amin(f_val):
      x_min=x_val[i]
      exit()

print('\nValue of x corresponding minimum value : ',x_min,'\n')

fig = plt.figure(figsize=(6,10))
# ax = plt.axes(projection='3d')

z = x**2 + y**2 + 2*x + 2*y

ax1 = fig.add_subplot(3,1,1,projection='3d')

ax1.plot_surface(x,y,z)
ax2 = fig.add_subplot(3,1,2,projection='3d')

ax2.contour3D(x,y,z,50)
ax3 = fig.add_subplot(3,1,3)
ax3.contour(x,y,z,50)
for i in range(len(x_val) - 1):
  plt.annotate('', xy=x_val[i+1,:], xytext=x_val[i,:],arrowprops={'arrowstyle': '-', 'color': 'r', 'lw': 1})

print(x_min[1])
plt.plot(x_min[0],x_min[1],'x',color='g')

plt.show()
```
